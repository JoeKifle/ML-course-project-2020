```python
# This function Splits dataset to train and test set
# Parameters
#     - dataSet : Type: pandas DataFrame, Desc: The original dataset containg all the data.
#     - testRation : Type: float, Desc: Number between 0-1. indicating percentage of testRatio. Eg. 0.2, 20% of the
#                    dataSet will be assigned to testData and the rest 80% will be training data.
def trainT_split(dataSet, testRation):
    np.random.seed(15)
    shuffledData = np.random.permutation(len(dataSet))
    testsetSize = int(len(dataSet)*testRation)
    testData = shuffledData[:testsetSize]
    trainData = shuffledData[testsetSize:]
    return dataSet.iloc[testData], dataSet.iloc[trainData]
```
